# CALIFORNIA HOUSE PRICE PREDICTION

## DESCRIPTION

This is a project that uses Data analysis libraries, and Supervised Machine Learning Libraries to predict prices of houses in the city of California.
Data was extracted from Kaggle.com.

Data Analysis Libraries: Numpy, Panadas, MatplotLib and Seaborn
The above libraries were used for exploration, Cleaning,visualization and preprocessing of the dataset.

Feature Engineering was performed for higher accuracy and the features were scaled as well

In the end we used 2 machine learning models during evaluation




## CODE

### EXPLORATION, VISUALISATION AND CLEANING OF DATA

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("housing.csv")
```

```python
data.head()
```

```python
data.info()
```

```python
data.dropna(inplace=True)   #This command makes all the non null values equal and saves the result in data object
```

```python
data.info()    #Now the data is clean
```

```python
from sklearn.model_selection import train_test_split

X = data.drop(['median_house_value'],axis = 1)              #all the values except the median_house_value i.e. drop
y = data['median_house_value']
```

```python
X_train, X_test, y_train , y_test = train_test_split(X,y,test_size=0.2)    #Train test split
```

```python
train_data = X_train.join(y_train)
```

```python
train_data
```

```python
train_data.hist(figsize = (15,8))            #Output shows that the data is skewed
```

```python
train_data.corr()               #Return correlation between data
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(train_data.corr(),annot = True, cmap="YlGnBu")
```

```python
train_data ['total_rooms'] = np.log(train_data['total_rooms'] + 1)
train_data['total_bedrooms'] = np. log(train_data['total_bedrooms']+1)
train_data['population'] = np. log(train_data['population']+1)
train_data['households'] = np. log(train_data['households']+1)               #To make the data less skewed
```

```python
train_data.hist(figsize=(15,8))             #data is less skewed
```

##### EXPLORING THE OCEAN PROXIMITY FEATURE

```python
train_data.ocean_proximity.value_counts()        #Groups the oceanproximity data and returns their count
```

```python
train_data = train_data.join(pd.get_dummies(train_data.ocean_proximity)).drop(['ocean_proximity'], axis=1)  
#This command first gets the sub features of ocean_proximity(Ex.<1H OCEAN) and then adds them to the train data. It then drops the ocean_proximity coloumn  
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(train_data.corr(),annot = True, cmap="YlGnBu")      
#shows that some features have different correlation(positive or negative) with ocean_proximity
```

```python
plt.figure(figsize= (15, 8))
sns.scatterplot(x="latitude", y="longitude", data = train_data, hue="median_house_value", palette="coolwarm")
#Plot shows that The median price on the coast is highest
```

### FEATURE ENGINEERING

```python
train_data['bedroom_ratio']  = train_data['total_bedrooms'] / train_data[ 'total_rooms']  #New feature of number of bedrooms/rooms
train_data['household_rooms']  = train_data['total_rooms'] / train_data[ 'households']  #New feature of number of rooms/households
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(train_data.corr(),annot = True, cmap="YlGnBu")   
```

## MACHINE LEARNING

### 1).LINEAR REGRESSION 

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train ,y_train = train_data.drop(['median_house_value'],axis=1),train_data['median_house_value']
X_train_s = scaler.fit_transform(X_train)

reg = LinearRegression()

reg.fit(X_train_s,y_train)
```

```python
test_data = X_test.join(y_test)

test_data ['total_rooms'] = np.log(test_data['total_rooms'] + 1)
test_data['total_bedrooms'] = np. log(test_data['total_bedrooms']+1)
test_data['population'] = np. log(test_data['population']+1)
test_data['households'] = np. log(test_data['households']+1)               

test_data = test_data.join(pd.get_dummies(test_data.ocean_proximity)).drop(['ocean_proximity'], axis=1)  

test_data['bedroom_ratio']  = test_data['total_bedrooms'] / test_data[ 'total_rooms']  
test_data['household_rooms']  = test_data['total_rooms'] / test_data[ 'households']
```

```python
X_test ,y_test = test_data.drop(['median_house_value'],axis=1),test_data['median_house_value']
```

```python
X_test_s = scaler.transform(X_test)
```

```python
reg.score(X_test_s,y_test)
```

### 2).RANDOM FOREST REGRESSOR

```python
from sklearn.ensemble import RandomForestRegressor

forest = RandomForestRegressor ()

forest.fit(X_train_s, y_train)
```

```python
forest.score(X_test_s,y_test)
```

```python
from sklearn.model_selection import GridSearchCV

forest = RandomForestRegressor()


param_grid = {
    "n_estimSeaborrrrrrators": [100,200,300],
    "min_samples_split": [2,4],
    "max_depth": [None,4,8]
}
grid_search = GridSearchCV(forest, param_grid, cv=5,
                        scoring="neg_mean_squared_error",
                        return_train_score=True)
grid_search.fit(X_train_s, y_train)
```

```python
grid_search.best_estimator_.score(X_test_s,y_test)
```

## CONCLUSION

We observed that Random Forest Regressor gave higher accuracy as compared to Linear Regressor.
